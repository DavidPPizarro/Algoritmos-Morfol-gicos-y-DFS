```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python

imagen = cv2.imread('komo.jpg', cv2.IMREAD_GRAYSCALE)
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

if imagen_rgb is None:
    print("Error: No se pudo cargar la imagen.")
else:
    print("Matriz de la imagen en RGB:")
    print(imagen_rgb) 
    
    # Mostrar una parte de la matriz (por si es muy grande)
    print("\nMatriz (primeros 10x10 p√≠xeles en RGB):")
    print(imagen_rgb[:100000, :100000, :])  # Muestra una secci√≥n de la matriz (R,G,B)

    # Mostrar la imagen
    plt.imshow(imagen_rgb)
    plt.title('Imagen en RGB')
    plt.axis('off')
    plt.show()
```

```python
kernel = np.ones((5,5), np.uint8)

erosion = cv2.erode(imagen_rgb, kernel, iterations=1)

cv2.imshow('Original', imagen)
cv2.imshow('Erosion', erosion)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
